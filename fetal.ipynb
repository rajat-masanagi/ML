```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
plt.pie(df['fetal_health'].value_counts(),labels=df['fetal_health'].value_counts().index)
plt.legend()
plt.title('1: Normal, 2: Suspect, 3: Pathological')
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print('Accuracy: ',accuracy_score(y_pred,y_test))
```

```python
fig = plt.figure(figsize=(10,6))
tree.plot_tree(model,feature_names=x.columns,filled=True)
plt.tight_layout()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=5)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print('Accuracy: ',accuracy_score(y_pred,y_test))
```

```python
fig = plt.figure(figsize=(10,6))
tree.plot_tree(model,feature_names=x.columns,filled=True)
plt.tight_layout()
```

```python
k=7
d=abs(df.corr()['fetal_health']).sort_values()[::-1][:k+1]
selected_col=d.index
```

```python
dd=df[selected_col]
```

```python
sns.heatmap(dd.corr(),annot=True)
plt.show()
```

```python
x=dd.iloc[:,1:]
y=dd.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(max_depth=5,random_state=2)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print('Accuracy: ',accuracy_score(y_pred,y_test))
```

```python
fig = plt.figure(figsize=(10,6))
tree.plot_tree(model,feature_names=x.columns,filled=True)
plt.tight_layout()
```
