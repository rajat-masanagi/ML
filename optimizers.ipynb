Sigmoid

```python
def sigmoid(y):
    return 1/(1+math.exp(-y))
```

Perceptron

```python
def neuron(x,w,b):
    yi=x*w+b
    return sigmoid(yi)
```

SGD

```python
def stochastic_gradient_descent(x,y,epochs=100,learning_rate=0.001):
    w=0
    b=0
    n=len(x)
    
    weights=[]
    mse=[]

    for i in range(epochs):
        total_error=0
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            dw=(yh-y[j])*yh*(1-yh)*x[j]
            db=(yh-y[j])*yh*(1-yh)
            
            w=w-dw*learning_rate
            b=b-db*learning_rate
        
            error=(yh-y[j])**2
            total_error+=error
            
        weights.append(w)
        mse.append(total_error)
        print(f'Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error/n:.4f}')
    
    return w,b
```

BGD

```python
def batch_gradient_descent(x,y,epochs=100,learning_rate=0.001):
    w=0
    b=0
    n=len(x)
    
    weights=[]
    mse=[]

    for i in range(epochs):
        total_error=0
        dw=0
        db=0
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            dw= dw + (yh-y[j])*yh*(1-yh)*x[j]
            db= db + (yh-y[j])*yh*(1-yh)
            
            error=(yh-y[j])**2
            total_error+=error
            
        w=w-dw*learning_rate/n
        b=b-db*learning_rate/n
                
        weights.append(w)
        mse.append(total_error)
        print(f'Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error:.4f}')
    
    return w,b
```

Mini BGD

```python
def MBGD(x,y,batch_size,epochs,lr):
    w=0
    b=0
    n=len(x)
    
    mse=[]
    
    for i in range(epochs):
        sample_no=0
        total_loss=0
        dw=0
        db=0
        
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            error=(yh-y[j])**2
            total_loss+=error
            
            dw=dw+(yh-y[j])*yh*(1-yh)*x[j]
            db=db+(yh-y[j])*yh*(1-yh)
            
            sample_no+=1
            if sample_no%batch_size==0:
                w=w-dw*lr
                b=b-db*lr
        
        print(f"Epoch {i}: MSE: {total_loss/n:.4f} Weight: {w:.4f} Bias: {b:.4f}")
        mse.append(total_loss)
        
    return w,b
        
```

Momentum GD

```python
def MGD(x,y,epochs=100,lr=0.001,beta=0.01):
    w=0
    b=0
    n=len(x)
    vb=1
    vw=1
    
    weights=[]
    mse=[]

    for i in range(epochs):
        total_error=0
        dw=0
        db=0
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            dw= dw + (yh-y[j])*yh*(1-yh)*x[j]
            db= db + (yh-y[j])*yh*(1-yh)
            
            error=(yh-y[j])**2
            total_error+=error
        
        vw=vw*beta*(1-beta)*dw
        vb=vb*beta*(1-beta)*db
        
        w=w+vw*lr
        b=b+vb*lr
                
        weights.append(w)
        mse.append(total_error/n)
        print(f'Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error/n:.4f}')
    
    return w,b
```

Adagrad

```python
def adagrad(x,y,epochs=100,lr=0.001,eps=1e-8):
    w=0
    b=0
    n=len(x)
    cw=0
    cb=0
    
    weights=[]
    mse=[]

    for i in range(epochs):
        total_error=0
        dw=0
        db=0
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            dw= dw + (yh-y[j])*yh*(1-yh)*x[j]
            db= db + (yh-y[j])*yh*(1-yh)
            
            error=(yh-y[j])**2
            total_error+=error
            
        cw+=dw**2
        cb+=db**2
        
        w=w-lr/(np.sqrt(cw)+eps)*dw
        b=b-lr/(np.sqrt(cb)+eps)*db
                
        weights.append(w)
        mse.append(total_error/n)
        print(f'Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error/n:.4f}')
    
    return w,b
```

RMSPROP

```python
def rmsprop(x,y,epochs=100,lr=0.001,eps=1e-8,beta=0.8):
    w=0
    b=0
    n=len(x)
    vw=1
    vb=1
    
    weights=[]
    mse=[]

    for i in range(epochs):
        total_error=0
        dw=0
        db=0
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            dw= dw + (yh-y[j])*yh*(1-yh)*x[j]
            db= db + (yh-y[j])*yh*(1-yh)
            
            error=(yh-y[j])**2
            total_error+=error
            
        vw=vw*beta+(1-beta)*dw**2
        vb=vb*beta+(1-beta)*db**2
        
        w=w-lr/(np.sqrt(vw)+eps)*dw
        b=b-lr/(np.sqrt(vb)+eps)*db
                
        weights.append(w)
        mse.append(total_error/n)
        print(f'Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error/n:.4f}')
    
    return w,b
```

Adadelta

```python
def adadelta(x,y,epochs=100,eps=1e-8,rho=0.95):
    w=0
    b=0
    n=len(x)
    cw=0
    cb=0
    adj_cw=0
    adj_cb=0
    
    weights=[]
    mse=[]

    for i in range(epochs):
        total_error=0
        dw=0
        db=0
        for j in range(n):
            yh=neuron(x[j],w,b)
            
            dw= dw + (yh-y[j])*yh*(1-yh)*x[j]
            db= db + (yh-y[j])*yh*(1-yh)
            
            error=(yh-y[j])**2
            total_error+=error
            
        cw=cw*rho + (1-rho)*dw**2
        cb=cb*rho + (1-rho)*db**2
        
        dwu=-(np.sqrt(adj_cw+eps)/np.sqrt(cw+eps))*dw
        dbu=-(np.sqrt(adj_cb+eps)/np.sqrt(cb+eps))*db
        
        adj_cw=rho*adj_cw+(1-rho)*dwu**2
        adj_cb=rho*adj_cb+(1-rho)*dbu**2
        
        w=w+dwu
        b=b+dbu
                
        weights.append(w)
        mse.append(total_error/n)
        print(f'Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error/n:.4f}')
    
    return w,b
```

Adam

```python
def adam(x, y, epochs=100, lr=0.001, eps=1e-8, beta1=0.9, beta2=0.999):
    w = 0
    b = 0
    n = len(x)
    mw = 0  # First moment for weights
    mb = 0  # First moment for biases
    vw = 0  # Second moment for weights
    vb = 0  # Second moment for biases

    weights = []
    mse = []

    for i in range(epochs):
        total_error = 0
        dw = 0
        db = 0

        for j in range(n):
            yh = neuron(x[j], w, b)
            
            dw += (yh - y[j]) * yh * (1 - yh) * x[j]
            db += (yh - y[j]) * yh * (1 - yh)

            error = (yh - y[j]) ** 2
            total_error += error

        # Update the first and second moments
        mw = beta1 * mw + (1 - beta1) * dw
        mb = beta1 * mb + (1 - beta1) * db

        vw = beta2 * vw + (1 - beta2) * dw**2
        vb = beta2 * vb + (1 - beta2) * db**2

        # Bias correction
        mw_corr = mw / (1 - beta1**(i + 1))  # Correct bias for first moment
        mb_corr = mb / (1 - beta1**(i + 1))
        
        vw_corr = vw / (1 - beta2**(i + 1))  # Correct bias for second moment
        vb_corr = vb / (1 - beta2**(i + 1))

        # Update weights and biases
        w = w - lr * mw_corr / (np.sqrt(vw_corr) + eps)
        b = b - lr * mb_corr / (np.sqrt(vb_corr) + eps)

        weights.append(w)
        mse.append(total_error / n)
        print(f"Epoch: {i+1}      Weight: {w:.4f}       Bias: {b:.4f}     MSE: {total_error/n:.4f}")

    return w, b
```
